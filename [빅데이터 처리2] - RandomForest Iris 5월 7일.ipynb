```python
from sklearn.datasets import load_iris
```

```python
load_iris()          #dictionary data. key :'data': 독립변수 &'target':종속변수 
```

```python
load_iris()   
```

```python
#독립변수,종속변수 출력  -> 보기 불편해서 dataframe으로도 봄 
```

```python
load_iris()["data"]   # 4칸 * 150줄
```

```python
import pandas as pd
pd.DataFrame(data=load_iris()["data"], 
             columns = ["sepal length","sepal width","petal length","petal width"])
```

```python
iris_df = pd.DataFrame(data=load_iris()["data"], 
             columns = ["sepal length","sepal width","petal length","petal width"])
```

```python
iris_df
```

```python
#종속변수 dataframe에 추가할껭 
```

```python
iris_df["target"] = load_iris()["target"]
```

```python
iris_df
```

```python
#loc은 줄칸 선택하는 함수 :은 모든줄 , 다음 칸조건인데 이상 이하 -> 다시 독립변수 선택함. 
```

```python
X = iris_df.loc[: , "sepal length":"petal width"]
X
```

```python
y = iris_df["target"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(X , y, test_size=0.2)
```

```python
# 0.2 로 test하고 0.8로 training
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RandomForestClassifier()        #디폴트 트리 개수: 100개
```

```python
rfc = RandomForestClassifier(oob_score = True)   
```

```python
rfc
```

```python
pd.DataFrame.from_records
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc.estimators_
```

```python
rfc.estimators_[99]
```
